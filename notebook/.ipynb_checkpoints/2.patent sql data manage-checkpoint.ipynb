# 处理detail为关系型数据

```python
detail_sql_df_filename = '../data/processed/detail_sql_df_0619.pkl'
if not os.path.isfile(detail_sql_df_filename):
    detail_data_keys = detail_df.loc[0,'data'].keys()
    detail_sql_df = detail_df[['pub_id','info_id']].copy()
    for k in detail_data_keys:
        detail_sql_df[k] = [i[k] for i in detail_df['data']]
    detail_sql_df.to_pickle(detail_sql_df_filename)
else:
    detail_sql_df = pd.read_pickle(detail_sql_df_filename)
```

# 处理legal为关系型数据

```python
legal_sql_df_filename = '../data/processed/legal_sql_df_0620.pkl'
if not os.path.isfile(legal_sql_df_filename):
    legal_data_df = legal_df[[True if 'data' in i else False for i in legal_df.data]].copy()

    legal_data_df = legal_data_df.reset_index(drop=True)

    legal_data_keys = list(legal_data_df.loc[0,'data']['data'][0].keys())

    legal_data_keys += ['pub_id','info_id']

    legal_data = []
    for num,i in enumerate(legal_data_df.itertuples()):
        if num %10000==0:
            print('[{0}] Num:{1}'.format(time.ctime(),num))
        _data = legal_data_df.loc[num,'data']['data']
        for d in _data:
            d['pub_id'] = legal_data_df.loc[num,'pub_id']
            d['info_id'] = legal_data_df.loc[num,'info_id']
            legal_data.append(d)

    legal_sql_df = pd.DataFrame()

    legal_sql_df['id'] = [i for i in range(len(legal_data))]
    for k in legal_data_keys:
        legal_sql_df[k] = [i[k] for i in legal_data]
else:
    legal_sql_df = pd.read_pickle(legal_sql_df_filename)
```

```python
cnx = mysql.connector.connect(user='root', password='ir@DUT#2018',
                              host='127.0.0.1',database='patentDB')
cursor = cnx.cursor()
```

```python
def detail_data_process(detail_tuple):
    _data = detail_tuple[2:]
    _data = ['|'.join(i) if isinstance(i,list) else i for i in _data]
    _format = [re.sub('[\'|\"]','','{0}'.format(i)) for i in _data]
    _format = ['"{0}"'.format(i) for i in _format]
    return ','.join(_format)
```

```python
for num,i in enumerate(detail_sql_df.itertuples()):
    if num%10000==0:
        print('[{0} Num:{1}]'.format(time.ctime(),num))
    base_sql_str = 'insert into detail_table ({0}) values({1})'
    columns_str = ','.join(list(detail_sql_df.columns)[1:])
    val_str = detail_data_process(i)
    sql_str = base_sql_str.format(columns_str,val_str)
#     print(num,sql_str)
    cursor.execute(sql_str)
```
