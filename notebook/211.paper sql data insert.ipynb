```python
def detail_data_process(detail_tuple):
    _data = detail_tuple[2:]
    _data = ['|'.join(i) if isinstance(i,list) else i for i in _data]
    _format = [re.sub('[\'|\"]','','{0}'.format(i)) for i in _data]
    _format = ['"{0}"'.format(i) for i in _format]
    return ','.join(_format)
```

```python
for num,i in enumerate(detail_sql_df.itertuples()):
    if num%10000==0:
        print('[{0} Num:{1}]'.format(time.ctime(),num))
    base_sql_str = 'insert into detail_table ({0}) values({1})'
    columns_str = ','.join(list(detail_sql_df.columns)[1:])
    val_str = detail_data_process(i)
    sql_str = base_sql_str.format(columns_str,val_str)
#     print(num,sql_str)
    cursor.execute(sql_str)
```
