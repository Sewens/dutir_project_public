```python
def dump_pickle(obj,fname):
    with open(fname,'wb') as file:
        pickle.dump(obj,file)
def load_pickle(fname):
    pickle.load(open(fname,'rb').read())
```

# 预处理一下

1、数据清洗（把里面的0001之类的删掉）

2、分词（专利法律词典）


```python
patent_word = [i.strip() for i in open('../data/user_data/patent_word_list.txt','r',encoding='utf-8').readlines()]
```

```python
patent_word = [i.split(' ')[-1] for i in patent_word]
```

```python
for word in patent_word:
    jieba.add_word(word)
```

```python
def wash_number(_str):
    pattern_num = re.compile('\d{4}\.')
    pattern_num1 = re.compile('\d{1}\.')
    _str = re.sub(pattern_num,'',_str)
    _str = re.sub(pattern_num1,'',_str)
    return _str
```

```python
claim_df['text'] = claim_df.text.apply(lambda x:wash_number(x))
```

```python
claim_df
```

```python
claim_df['seg_text'] = claim_df.text.apply(jieba.lcut)
```

```python
claim_df.to_pickle('../data/processed/claim_df_seg_0630.pkl')
```

# 向量化 使用tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
patent_text = claim_df.seg_text.tolist()
patent_text = [' '.join(i) for i in patent_text]
```

```python
patent_tfidf_obj = TfidfVectorizer()
```

```python
patent_tfidf_obj.fit(patent_text)
```

```python
dump_pickle(patent_tfidf_obj,'../data/processed/patent_tfidf_obj_default_0701.pkl')
```

```python
patent_all_tfidf_matrix = patent_tfidf_obj.transform(patent_text)
```

# PCA 降维对象

```python
from sklearn.decomposition import TruncatedSVD
```

```python
pca_obj = TruncatedSVD(n_components=250)
```

```python
%timeit
pca_obj.fit(patent_all_tfidf_matrix)
```

```python
dump_pickle(pca_obj,'../data/processed/patent_pca_obj_default_0701.pkl')
```

# LDA 降维对象

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
lda_obj = LatentDirichletAllocation(n_components=100,)
```

```python
lda_obj.fit(patent_all_tfidf_matrix)
```

```python
dump_pickle(lda_obj,'../data/processed/patent_lda_obj_default_0701.pkl')
```
