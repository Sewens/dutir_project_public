```python
def dump_pickle(obj,fname):
    with open(fname,'wb') as file:
        pickle.dump(obj,file)
def load_pickle(fname):
    pickle.load(open(fname,'rb').read())
```

```python
claim_df = pd.read_pickle('../data/patent_data/processed/claim_df_seg_0630.pkl')
```

```python
claim_df
```

# 向量化 使用tfidf 限定1w个词

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
patent_text = claim_df.seg_text.tolist()
patent_text = [' '.join(i) for i in patent_text]
```

```python
patent_tfidf_obj = TfidfVectorizer(max_features=10000)
```

```python
patent_tfidf_obj.fit(patent_text)
```

```python
dump_pickle(patent_tfidf_obj,'../data/patent_data/model/patent_tfidf_model.pkl')
```

```python
patent_all_tfidf_matrix = patent_tfidf_obj.transform(patent_text)
```

# PCA 降维对象

```python
from sklearn.decomposition import TruncatedSVD
```

```python
pca_obj = TruncatedSVD(n_components=250)
```

```python
pca_obj.fit(patent_all_tfidf_matrix)
```

# LDA 降维对象

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
baseNmaeTemplate = '../data/patent_data/model/patent_lda_topic{0}_model.pkl'
```

```python
for i in range(2,10):
    print('Model LDA topic:',i)
    lda_obj = LatentDirichletAllocation(n_components=i,)
    lda_obj.fit(patent_all_tfidf_matrix)
    dump_pickle(lda_obj,baseNmaeTemplate.format(i))
```

```python
lda_obj = LatentDirichletAllocation(n_components=100,n_jobs=12)
lda_obj.fit(patent_all_tfidf_matrix)
dump_pickle(lda_obj,baseNmaeTemplate.format(100))
```

# 数据准备

```python
sample_patent_data = patent_tfidf_obj.transform(patent_text[:10000])
```

```python
sample_pca_data = pca_obj.transform(sample_patent_data)
```

```python
sample_lda_data = lda_obj.transform(sample_patent_data)
```

# 聚类 使用 k-means

```python
from sklearn.cluster import KMeans
```

```python
km_tfidf = KMeans(n_clusters=8)
```

```python
km_pca = KMeans(n_clusters=8)
```

```python
km_lda = KMeans(n_clusters=8)
```

```python
cluster_tfidf = km_tfidf.fit_predict(sample_patent_data)
```

```python
cluster_pca = km_pca.fit_predict(sample_pca_data)
```

```python
cluster_lda = km_lda.fit_predict(sample_lda_data)
```
